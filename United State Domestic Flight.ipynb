```python
import numpy as np
import pandas as pd
```

```python
#make data frame from csv file
data_flight = pd.read_csv (r'C:\Users\ZELIEVES\Downloads\United State Domestic Flights.csv')
data_flight [['flight_date','unique_carrier','flight_num','origin','dest','arr_delay','cancelled','distance','carrier_delay','weather_delay','late_aircraft_delay','nas_delay','security_delay','actual_elapsed_time']]
```

```python
#clean dataset by fill missing data with nan
clean_data_flight = data_flight.fillna(np.nan)
data_flight.head()
```

```python
#sort data frame from arr_delay in descending order 
sort_data = clean_data_flight.sort_values("arr_delay", axis = 0, ascending = False, inplace = False, na_position ='last') 
sort_data.head()
```

```python
# use lambda function to sort occuring (flight_delay) that records True or False 
data_flight_delay = sort_data ["arr_delay"].apply(lambda f: f >0)
print(data_flight_delay)
```

```python
#use value count to sort aggregate trends of flight_delay
count_sort_data = data_flight_delay. value_counts()
print(count_sort_data)
```

```python
#count total percentage of flight_delay
flight_not_delayed = data_flight_delay .value_counts()[0]
data_flight_delayed = data_flight_delay .value_counts()[1]
total_sum_flight = flight_not_delayed + data_flight_delayed
#calculate %
print (float (data_flight_delayed)/total_sum_flight)
```

```python
#calculate numbers of fight_delay longer than 20 minutes
sort_data ['longer_than_20mins'] = sort_data ["arr_delay"]. apply(lambda f: f >20)
count_delay_20mins = sort_data ['longer_than_20mins'].value_counts()
print(count_delay_20mins)
```
